```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
total=purchase_data["SN"].nunique()
total_players=pd.DataFrame({
    "Total Players": [total],
})
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Number_Unique_Items=purchase_data["Item ID"].nunique()
# print(Number_Unique_Items)
Average_Price="${:,.2f}".format(round(purchase_data["Price"].mean(),2))
# print(Average_Price)
Number_of_Purchases=purchase_data["Purchase ID"].count()
# print(Number_of_Purchases)
Total_Revenue="${:,.2f}".format(purchase_data["Price"].sum())
# print(Total_Revenue)
purchase_analysis=pd.DataFrame({
    "Number of Unique Items": [Number_Unique_Items],
    "Average Price": [Average_Price],
    "Number of Purchases": [Number_of_Purchases],
    "Total_Revenue": [Total_Revenue]
})
purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
purchase_sorted=purchase_data.sort_values(by=['SN'])
purchase_dropped=purchase_sorted.drop_duplicates(['SN'], keep='last')
count=purchase_dropped.groupby(["Gender"]).count()

total=purchase_dropped["SN"].nunique()
total
data=pd.DataFrame({
    "Total Count": count["SN"],
    "Percentage of Players": (round(count["SN"]/total,4))
})
data['Percentage of Players'] = data['Percentage of Players'].astype(float).map("{:.2%}".format)
data
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
grouped=purchase_data.groupby(["Gender"])
count2=purchase_data.groupby(["Gender"]).count()
average=purchase_data.groupby(["Gender"])["Price"].mean()
purchase_value=purchase_data.groupby(["Gender"])["Price"].sum()


data2=pd.DataFrame({
    "Purchase Count": count2["SN"],
    "Average Purchase Price": round(average,2),
    "Total Purchase Value": purchase_value,
    "Avg Total Purchase per Person": round(average,2)
})
data2['Average Purchase Price'] = data2['Average Purchase Price'].astype(float).map('${:,.2f}'.format)
data2['Total Purchase Value'] = data2['Total Purchase Value'].astype(float).map('${:,.2f}'.format)
data2['Avg Total Purchase per Person'] = data2['Avg Total Purchase per Person'].astype(float).map('${:,.2f}'.format)
data2
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 200]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_age_summary=purchase_data

purchase_age_summary=purchase_age_summary.sort_values(by=['SN'])
purchase_age_summary=purchase_age_summary.drop_duplicates(['SN'], keep='last')

purchase_age_summary["Age Summary"] = pd.cut(purchase_age_summary["Age"], bins, labels=group_labels)
purchase_age_summary_grouped=purchase_age_summary.groupby("Age Summary")
purchase_age_summary_grouped = pd.DataFrame(purchase_age_summary_grouped.count())
purchase_age_summary_grouped["Percentage of Players"] = (purchase_age_summary_grouped["Purchase ID"]/total)*100
purchase_age_summary_grouped = purchase_age_summary_grouped[["SN","Percentage of Players"]]
purchase_age_summary_grouped = purchase_age_summary_grouped.rename(columns={"SN":"Total Count"})
purchase_age_summary_grouped["Percentage of Players"] = purchase_age_summary_grouped["Percentage of Players"].map("{:,.2f}%".format)
purchase_age_summary_grouped

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_total=purchase_data
purchase_total["Age Summary"] = pd.cut(purchase_total["Age"], bins, labels=group_labels)
purchase_total


pgroup=purchase_total.groupby("Age Summary")["Price"].count()
purchase_summary = pd.DataFrame(pgroup)
purchase_summary = purchase_summary.rename(columns={"Price":"Purchase Count"})
pgroup=purchase_total.groupby("Age Summary")["Price"].mean()
purchase_summary["Average Purchase Price"] = round(pd.DataFrame(pgroup),2)
pgroup=purchase_total.groupby("Age Summary")["Price"].sum()
purchase_summary["Total Purchase Value"] = pd.DataFrame(pgroup)
purchase_summary["Avg Total Purchase per Person"] = round(purchase_summary["Total Purchase Value"]/purchase_summary["Purchase Count"],2)
purchase_summary['Average Purchase Price'] = purchase_summary['Average Purchase Price'].astype(float).map('${:,.2f}'.format)
purchase_summary['Total Purchase Value'] = purchase_summary['Total Purchase Value'].astype(float).map('${:,.2f}'.format)
purchase_summary['Avg Total Purchase per Person'] = purchase_summary['Avg Total Purchase per Person'].astype(float).map('${:,.2f}'.format)

purchase_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_total=purchase_data
SNgroup=purchase_total.groupby("SN")["Purchase ID"].count()
SN = pd.DataFrame(SNgroup)
SN = SN.rename(columns={"Purchase ID":"Purchase Count"})
SNgroup=round(purchase_total.groupby("SN")["Price"].mean(),2)
SN["Average Purchase"]= SNgroup
SNgroup=purchase_total.groupby("SN")["Price"].sum()
SN["Total Purchase Value"]= SNgroup
SN.sort_values(by=['Purchase Count'], inplace=True, ascending=False)
SN['Average Purchase'] = SN['Average Purchase'].astype(float).map('${:,.2f}'.format)
SN['Total Purchase Value'] = SN['Total Purchase Value'].astype(float).map('${:,.2f}'.format)

SN.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_total=purchase_data
ItemGroup=purchase_total.groupby("Item Name")["Item ID"].mean().astype(int)
Item=ItemGroup
Item = pd.DataFrame(ItemGroup)
ItemGroup=purchase_total.groupby("Item Name")["Purchase ID"].count()
Item["Purchase Count"] = pd.DataFrame(ItemGroup)
ItemGroup=round(purchase_total.groupby("Item Name")["Price"].mean(),2)
Item["Item Price"] = pd.DataFrame(ItemGroup)
ItemGroup=purchase_total.groupby("Item Name")["Price"].sum()
Item["Total Purchase Value"] = pd.DataFrame(ItemGroup)
Item.sort_values(by=['Purchase Count'], inplace=True, ascending=False)
Item['Item Price'] = Item['Item Price'].astype(float).map('${:,.2f}'.format)
#Item['Total Purchase Value'] = Item['Total Purchase Value'].astype(float).map('${:,.2f}'.format)
Item.head(5)
```
